# Graph Generator

This notebook contains cells to make pyplot graphs of all of the data we have.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Regression Models

### Decision Tree Regressor

```python
# min_samples_split
dt_mss_df = pd.read_csv('data/MTGDecisionTreeRegressorMinSamplesSplit.csv')
dt_mss_x = dt_mss_df['min_samples_split'].to_numpy()
dt_mss_y1 = dt_mss_df['Average Training Score'].to_numpy()
dt_mss_y2 = dt_mss_df['Average Test Score'].to_numpy()
dt_mss_y3 = dt_mss_df['Mean Absolute Error'].to_numpy()
dt_mss_y4 = dt_mss_df['Mean Squared Error'].to_numpy()

figure, axis = plt.subplots(2, 2, figsize=(12, 12))
figure.dpi = 150

axis[0, 0].plot(dt_mss_x, dt_mss_y1)
axis[0, 0].set_title('Average Training Accuracy')

axis[0, 1].plot(dt_mss_x, dt_mss_y2)
axis[0, 1].set_title('Average Test Accuracy')

axis[1, 0].plot(dt_mss_x, dt_mss_y3)
axis[1, 0].set_title('Mean Absolute Error')

axis[1, 1].plot(dt_mss_x, dt_mss_y4)
axis[1, 1].set_title('Mean Squared Error')

plt.savefig('images/dtr_mss.png')
plt.show()
```

```python
# criterion
dt_cri_df = pd.read_csv('data/MTGDecisionTreeRegressorCriterion.csv')
dt_cri_x = dt_cri_df['Activation Function'].to_numpy()
dt_cri_y1 = dt_cri_df['Average Training Accuracy'].to_numpy()
dt_cri_y2 = dt_cri_df['Average Test Accuracy'].to_numpy()
dt_cri_y3 = dt_cri_df['Mean Absolute Error'].to_numpy()
dt_cri_y4 = dt_cri_df['Mean Squared Error'].to_numpy()

figure, axis = plt.subplots(2, 2, figsize=(12, 12))
figure.dpi = 150

axis[0, 0].bar(dt_cri_x, dt_cri_y1)
axis[0, 0].set_title('Average Training Accuracy')
axis[0, 0].set_ylim(0.75, 1.01)

axis[0, 1].bar(dt_cri_x, dt_cri_y2)
axis[0, 1].set_title('Average Test Accuracy')
axis[0, 1].set_ylim(0.75, 0.801)

axis[1, 0].bar(dt_cri_x, dt_cri_y3)
axis[1, 0].set_title('Mean Absolute Error')
axis[1, 0].set_ylim(120, 130)

axis[1, 1].bar(dt_cri_x, dt_cri_y4)
axis[1, 1].set_title('Mean Squared Error')
axis[1, 1].set_ylim(135000, 145000)

plt.savefig('images/dtr_cri.png')
plt.show()
```

```python
# pruning
dt_pru_df = pd.read_csv('data/MTGDecisionTreeRegressorPruning.csv')
dt_pru_x = dt_pru_df['ccp_aplpha'].to_numpy()
dt_pru_y1 = dt_pru_df['Average Training Accuracy'].to_numpy()
dt_pru_y2 = dt_pru_df['Average Test Accuracy'].to_numpy()
dt_pru_y3 = dt_pru_df['Mean Absolute Error'].to_numpy()
dt_pru_y4 = dt_pru_df['Mean Squared Error'].to_numpy()

figure, axis = plt.subplots(2, 2, figsize=(12, 12))
figure.dpi = 150

axis[0, 0].plot(dt_pru_x, dt_pru_y1)
axis[0, 0].set_title('Average Training Accuracy')

axis[0, 1].plot(dt_pru_x, dt_pru_y2)
axis[0, 1].set_title('Average Test Accuracy')

axis[1, 0].plot(dt_pru_x, dt_pru_y3)
axis[1, 0].set_title('Mean Absolute Error')

axis[1, 1].plot(dt_pru_x, dt_pru_y4)
axis[1, 1].set_title('Mean Squared Error')

plt.savefig('images/dtr_pru.png')
plt.show()
```
